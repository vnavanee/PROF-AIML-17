# social and economic context attributes
16 - emp.var.rate: employment variation rate - quarterly indicator (numeric)
17 - cons.price.idx: consumer price index - monthly indicator (numeric)
18 - cons.conf.idx: consumer confidence index - monthly indicator (numeric)
19 - euribor3m: euribor 3 month rate - daily indicator (numeric)

### Problem 4: Understanding the Task

After examining the description and data, your goal now is to clearly state the *Business Objective* of the task.  State the objective below.

Business Background: Due to increased competition European banks are looking to increase financial assets they hold. A portuguese bank from where this data comes from used multiple direct marketing campaigns to have customers sign up for a long term deposit product by offering attractive incentives. These campaigns having an inherent risk of alienating the customers that may get annoyed by marketing phone calls which needs to be balanced with the need to attract them to a new product.

Background Data: Data was collected from a Portugues bank that had conducted several contact-center based direct marketing campaigns. This was accomplished mostly using telephone and to a lesser extent using online banking channel. This data spanned between May 2008 and November 2010. This data represents about 17 campaigns with 6499 rows and a success rate of 8% making it an imbalanced classes data.

Business Objective: To determine a profile of a bank customer who will subscribe or sign up for a term deposit so that the marketing campaigns are designed to efficiently attract the right customers and minimize customer dissatisfaction.

### Problem 5: Engineering Features

Now that you understand your business objective, we will build a basic model to get started.  Before we can do this, we must work to encode the data.  Using just the bank information features (columns 1 - 7), prepare the features and target column for modeling with appropriate encoding and transformations.

````python
X = df.iloc[:, [0,1,2,3,4,5,6]]
y = df.iloc[:,-1]
````

````python
# select all columns with a data type of 'object'
selector = make_column_selector(dtype_include=object)
# transformer for all models except SVC which does require PolynomialFeatures - where it is implicity done.
transformer = make_column_transformer((OneHotEncoder(drop = 'if_binary'), selector),
                                     (PolynomialFeatures(degree = 2), ['age']))
# transformer without polynomial feature engineering
transformer_nopoly = make_column_transformer((OneHotEncoder(drop = 'if_binary'), selector))
#transformer for SVC

svc_transformer = make_column_transformer((OneHotEncoder(drop = 'if_binary'), selector), remainder = StandardScaler())
                                    
````

### Problem 6: Train/Test Split

With your data prepared, split it into a train and test set.

````python
X_train, X_test, y_train, y_test = train_test_split(X, y)
````

````python
lgr_pipe = Pipeline([('transformer', transformer),
                               ('lgr', LogisticRegression(random_state=42, max_iter = 1000))])
lgr_pipe.fit(X_train, y_train)
````

````python
imp = permutation_importance(lgr_pipe, X_train, y_train)
imp_cols = zip(X_train.columns,imp.importances_mean)
#print(imp.importances_mean)
model_1_df = pd.DataFrame(imp.importances_mean, columns=['Features'],index=X_train.columns)
print(list(imp_cols))
model_1_df.plot.barh()
````

````python
coef_df = pd.DataFrame(lgr_pipe[-1].coef_[0], columns=['Features'], index=lgr_pipe[0].get_feature_names_out())
print(coef_df)
coef_df.plot.barh(figsize=(10,10))
````

Job and Education seem to be highly and positively correlated with the outcome. So are retired customer, students, illiterate customers and customers who have not defaulted before.

#### Results & Metrics change - Precision and Threshold

````python
def precision_thresh(predict_probs, 
                  y_test, 
                  thresh):
    preds = np.where(predict_probs >= thresh, 'yes', 'no')
    
    return precision_score(y_test, preds, pos_label='yes')
````

````python
def recall_thresh(predict_probs, 
                  y_test, 
                  thresh):
    preds = np.where(predict_probs >= thresh, 'yes', 'no')
    
    return recall_score(y_test, preds, pos_label='yes')
````

````python
results_df = pd.DataFrame({'threshold': thresholds, 'precision': [precision_thresh(test_predictions[:, 1], y_test, i) for i in thresholds]})
results_df['recall'] = [recall_thresh(test_predictions[:, 1], y_test, i) for i in thresholds]
plt.plot(results_df['threshold'], results_df['precision'],'--o',label='Precision')
plt.plot(results_df['threshold'], results_df['recall'], '--o',label='Recall')
plt.grid()
plt.ylabel('Threshold')
plt.xlabel('Score')
plt.title('Precison and Recall Vs Decision Thresholds')
plt.legend(loc='center right')
````

````python
fpr, tpr, thresh_rocs = roc_curve(y_test, test_predictions[:,1], pos_label='yes')
print('False Positive Rates: ', fpr[:5])
print('True Positive Rates: ', tpr[:5])
plt.plot(fpr, tpr, '--o', label = 'roc curve')
plt.plot(tpr, tpr, label = 'baseline')
plt.legend()
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.grid();
````
